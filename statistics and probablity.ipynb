```python
for i in range(3):
    print(f"Outer loop iteration (for loop): i = {i}")
    j = 0
    while j < 2:
        print(f"  Inner loop iteration (while loop): j = {j}")
        j += 1
    print("\n") # Add a newline for better readability between outer loop iterations
```

In this example:

*   The **`for i in range(3)`** loop is the outer loop. It will iterate three times, with `i` taking values `0`, `1`, and `2`.
*   For each iteration of the outer `for` loop, the **`while j < 2`** loop (the inner loop) starts fresh.
*   Inside the inner `while` loop, `j` is initialized to `0` and then increments by `1` in each of its iterations. The `while` loop continues as long as `j` is less than `2`.

This means the inner `while` loop will execute twice for each iteration of the outer `for` loop, resulting in a total of `3 * 2 = 6` executions of the inner loop's body.
