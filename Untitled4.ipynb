```python
def predict_yield (rainfall, temperature, soil_quality):
    basic_yield = 40  # quintals per hectare

    # Rainfall adjustment
    rainfall_change = 0
    if rainfall > 700:
        rainfall_change = 2 * ((rainfall - 700) / 10)
    elif rainfall < 300:
        rainfall_change = -2 * ((300 - rainfall) / 10)

    # Temperature adjustment
    temperature_change = 0
    if temperature < 20:
        temperature_change = -(25 - temperature)
    elif temperature > 40:
        temperature_change = -(temperature - 30)

    # Soil quality factor
    soil_factor = soil_quality / 100

    # Final yield calculation
    final_yield = basic_yield * (1 + rainfall_change / 100) \
                              * (1 + temperature_change / 100) \
                              * soil_factor

    return round(final_yield, 2)


# Example usage
rainfall = float(input("Enter rainfall (mm): "))
temperature = float(input("Enter temperature (°C): "))
soil_quality = float(input("Enter soil quality (80-100): "))

yield_prediction = predict_yield(rainfall, temperature, soil_quality)
print("Predicted Crop Yield:", yield_prediction, "quintals/hectare")
```
